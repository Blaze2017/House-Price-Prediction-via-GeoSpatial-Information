## Visulizing spatial data with Python

```python
%matplotlib inline

import pandas as pd
import geopandas

import matplotlib.pyplot as plt

pd.options.display.max_rows = 10
```

```python
countries = geopandas.read_file("/Users/blaze/OneDrive/Work/GeoSpatial/ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp")
cities = geopandas.read_file("/Users/blaze/OneDrive/Work/GeoSpatial/ne_110m_populated_places/ne_110m_populated_places.shp")
rivers = geopandas.read_file("/Users/blaze/OneDrive/Work/GeoSpatial/ne_50m_rivers_lake_centerlines_scale_rank/ne_50m_rivers_lake_centerlines_scale_rank.shp")
```

## Commonly used functions

```python
countries.plot()
```

```python
countries.plot(figsize=(20, 20))
```

```python
ax = countries.plot(figsize=(15, 15))
ax.set_axis_off()
```

We can introducing the colored distribution into the map.

```python
countries = countries[(countries['POP_EST'] >0 ) & (countries['NAME'] != "Antarctica")]
```

```python
countries['GDP_PER_CAP'] = countries['GDP_MD_EST'] / countries['POP_EST'] * 100
```

Color the map based on the column 'GDP_PER_CAP'

```python
ax = countries.plot(figsize=(15, 15), column='GDP_PER_CAP')
ax.set_axis_off()
```

Add the legend on:

```python
ax = countries.plot(figsize=(15, 15), column='GDP_PER_CAP', scheme = 'quantiles', legend = True)
ax.set_axis_off()
```

#### Combining different dataframe on a single plot

```python
ax = countries.plot(figsize=(15, 15))
cities.plot(ax=ax, color='red', markersize=10)
ax.set_axis_off()
```

```python
ax = countries.plot(edgecolor='k', facecolor='none', figsize=(15, 10))
rivers.plot(ax=ax)
cities.plot(ax=ax, color='C1')
ax.set(xlim=(60, 160), ylim=(20, 70))
```
